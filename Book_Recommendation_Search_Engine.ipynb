```python
import gzip
#to view the contents of single line
with gzip.open("goodreads_books.json.gz",'r') as f:
  line=f.readline()
```

```python
line
```

```python
import json
#changes the line to python dictionary
json.loads(line)
```

```python
def parse_fields(line):
  data=json.loads(line)
  return{
      "book_id":data["book_id"],
      "title": data["title_without_series"],
      "ratings":data["ratings_count"],
      "url":data["url"],
      "cover_image":data["image_url"]
  }
```

```python
books_titles=[]
with gzip.open("goodreads_books.json.gz",'r') as f:
  while True:
    line=f.readline()
    if not line:
      break
    fields=parse_fields(line)
#read line by line
    try:
      ratings=int(fields["ratings"])
    except ValueError:
      continue
#if no. of ratings > 25 then add it to the list
    if ratings>25:
      books_titles.append(fields)
```

```python
import pandas as pd
#convert it to dataframe
titles=pd.DataFrame.from_dict(books_titles)
```

```python
titles["ratings"]=pd.to_numeric(titles["ratings"])
```

```python
titles["mod_title"]=titles["title"].str.replace("[^a-zA-Z0-9 ]","",regex=True)
```

```python
titles
```

```python
titles["mod_title"]=titles["mod_title"].str.lower()
```

```python
#Remove spaces that has more spaces
titles["mod_title"]=titles["mod_title"].str.replace("\s+"," ",regex=True)
```

```python
titles
```

```python
titles=titles[titles["mod_title"].str.len()>0]
#if after all the modifications in title, if the title became empty string, then remove the row
```

```python
titles.to_json("book_titles.json")
#store it as json file
```

**Generating search engine**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
tfidf=vectorizer.fit_transform(titles["mod_title"])
#to generate frequency matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity

import numpy as np
import re

def show_image(val):
  return '<img src="{}" width=50></img>'.format(val)

def search(query,vectorizer):
  processed=re.sub("[^a-zA-Z0-9 ]","",query.lower())
  query_vec=vectorizer.transform([processed])
  similarity=cosine_similarity(query_vec,tfidf).flatten()
  #how much each row in a matrix is similar to the given title
  indices=np.argpartition(similarity, -10)[-10:]
  #finding 10 largest similarity values.(i.e) indices of those values
  results=titles.iloc[indices]
  #take results that has more ratings
  results=results.sort_values("ratings",ascending=False)
  return results.head(5).style.format({'cover_image':show_image})
```

```python
search("The Girl Who Played with Fire",vectorizer)
```
